```python
import pandas as pd

data = pd.read_csv('../../_git/donnees/meteo/2023/synop.202309.csv', sep=';', dtype={'date':'str'}, na_values='mq')
data = data.rename(columns={
    'numer_sta': 'Station', 
    'date': 'DateHeure', 
    'dd' : 'DirectionVent', 
    'ff' : 'VitesseVent', 
    't': 'Temperature', 
    'u' : 'Humidite', 
    'vv' : 'Visibilite', 
    'pres' : 'Pression', 
    'rr1' : 'Precipitation01',
    'rr3' : 'Precipitation03',
    'rr6' : 'Precipitation06',
    'rr12' : 'Precipitation12',
    'rr24': 'Precipitation24',
})
#data.info()
data.drop(columns=['Unnamed: 59'],inplace=True)
#data.replace('mq', 'NaN', inplace=True)

data.set_index('Station')
data['DateHeure'] = pd.to_datetime(data["DateHeure"], format="mixed")

data['Temperature'] = data['Temperature'] - 273.15

data['Mois']      = data.DateHeure.dt.month
data['Annee']     = data.DateHeure.dt.year
data['AnneeMois'] = data.DateHeure.dt.year*100 + data.DateHeure.dt.month
data['Semaine']   = data.DateHeure.dt.isocalendar().week
data['MoisJour']  = data.DateHeure.dt.month*100 + data.DateHeure.dt.day
data['JourA']     = data.DateHeure.dt.dayofyear

#data.reset_index(inplace=True)
data.tail()
```

```python
# Aggregation

df = data[['Temperature', 'Pression', 'JourA']].groupby(['JourA']).agg({'Temperature': 'mean', 'Pression': ['min', 'max', 'mean']})
df
```
