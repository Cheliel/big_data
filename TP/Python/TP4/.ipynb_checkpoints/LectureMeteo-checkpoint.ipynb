```python
import pandas
```

# Lecture du fichier Synop.202309

```python
donnees = pandas.read_csv('donnees/synop.202309.csv',
                           sep=';',
                           usecols=['numer_sta','date','dd','ff','t','u','vv',
                                    'pres','rr1','rr3','rr6','rr12','rr24'],
                           na_values='mq',
                           dtype={'numer_sta':str,'date':str},
                          ).rename( columns={'numer_sta':'Station',
                                      'date':'DateHeure',
                                      'dd':'DirectionVent',
                                      'ff':'VitesseVent',
                                      't':'Temperature',
                                      'u':'Humidite',
                                      'vv':'Visibilite',
                                      'pres':'Pression',
                                      'rr1' :'Precipitation01',
                                      'rr3' :'Precipitation03',
                                      'rr6' :'Precipitation06',
                                      'rr12':'Precipitation12',
                                      'rr24':'Precipitation24'})
```

```python
donnees.info()
```

```python
donnees.head()
```

```python
donnees["DateHeure"] = pandas.to_datetime(donnees["DateHeure"], format='%Y%m%d%H%M%S')
donnees.Temperature  = donnees.Temperature - 273.15
donnees.Pression     = donnees.Pression / 100
donnees.Visibilite   = donnees.Visibilite / 1000
```

```python
donnees['Mois']      = donnees.DateHeure.dt.month
donnees['Annee']     = donnees.DateHeure.dt.year
donnees['AnneeMois'] = donnees.DateHeure.dt.year*100 + donnees.DateHeure.dt.month
donnees['Semaine']   = donnees.DateHeure.dt.isocalendar().week
donnees['MoisJour']  = donnees.DateHeure.dt.month*100 + donnees.DateHeure.dt.day
donnees['JourA']     = donnees.DateHeure.dt.dayofyear
```

```python
donnees.tail()
```

```python
donnees.set_index(['Station','DateHeure'],inplace=True)
```
