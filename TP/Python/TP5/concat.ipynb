```python
import pandas as pd
from os import listdir
import glob


def getDirectoryList(directoryPath="../../_git/donnees/meteo/2023"):
    return glob.glob(glob.escape(directoryPath) + "/*.csv")

def getDataFrames(path="../../_git/donnees/meteo/2023"):
    filesPath = getDirectoryList()
    DFlist = []
    for i in range((len(filesPath))):
        DFlist.append(getDF(filesPath[i]))
    return pd.concat(DFlist, ignore_index=True)


def getDF(path):
    data = pd.read_csv(path, sep=';', dtype={'date':'str'}, na_values='mq')
    data.rename(columns={
        'numer_sta': 'Station', 
        'date': 'DateHeure', 
        'dd' : 'DirectionVent', 
        'ff' : 'VitesseVent', 
        't': 'Temperature', 
        'u' : 'Humidite', 
        'vv' : 'Visibilite', 
        'pres' : 'Pression', 
        'rr1' : 'Precipitation01',
        'rr3' : 'Precipitation03',
        'rr6' : 'Precipitation06',
        'rr12' : 'Precipitation12',
        'rr24': 'Precipitation24',
    })
    return data



        
    
        
data = getDataFrames()

data
#data = pd.read_csv('../../_git/donnees/meteo/2023/synop.202309.csv', sep=';', dtype={'date':'str'}, na_values='mq')

```
