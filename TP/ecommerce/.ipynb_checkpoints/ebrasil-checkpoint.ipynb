## Lecture des fichiers sur le dique 

```python


olist_orders = pd.read_csv('../../_git/donnees/ebrasil/olist_orders_dataset.csv',
                           sep=',',
                          # usecols=['numer_sta','date','dd','ff','t','u','vv',
                          #         'pres','rr1','rr3','rr6','rr12','rr24'],
                           na_values='mq',
                           dtype={},
                          )

olist_orders.order_purchase_timestamp = pd.to_datetime(olist_orders.order_purchase_timestamp)
olist_orders.order_approved_at = pd.to_datetime(olist_orders.order_approved_at)
olist_orders.order_delivered_carrier_date = pd.to_datetime(olist_orders.order_delivered_carrier_date)
olist_orders.order_delivered_customer_date = pd.to_datetime(olist_orders.order_delivered_customer_date)
olist_orders.order_estimated_delivery_date   = pd.to_datetime(olist_orders.order_estimated_delivery_date )

olist_items = pd.read_csv('../../_git/donnees/ebrasil/olist_order_items_dataset.csv',
                           sep=',',
                          # usecols=['numer_sta','date','dd','ff','t','u','vv',
                          #         'pres','rr1','rr3','rr6','rr12','rr24'],
                           na_values='mq',
                           dtype={'numer_sta':str,'date':str},
                          )

olist_items.shipping_limit_date    = pd.to_datetime(olist_items.shipping_limit_date )
 
olist_payments = pd.read_csv('../../_git/donnees/ebrasil/olist_order_payments_dataset.csv',
                           sep=',',
                          # usecols=['numer_sta','date','dd','ff','t','u','vv',
                          #         'pres','rr1','rr3','rr6','rr12','rr24'],
                           na_values='mq',
                           dtype={'numer_sta':str,'date':str},
                          )





```

## Liste des commandes + items commandé

```python
commandes = olist_orders.merge(olist_items, on="order_id")
commandes.head()
```

## DataFrame - to parquet 

```python
olist_orders.to_parquet('../../_git/donnees/TP/ebrasil/orders.gzip',compression='gzip', engine='pyarrow')
olist_items.to_parquet('../../_git/donnees/TP/ebrasil/items.gzip',compression='gzip', engine='pyarrow')
olist_payments.to_parquet('../../_git/donnees/TP/ebrasil/payments.gzip',compression='gzip', engine='pyarrow')
```

```python
# Sanatize 
# pivot_table => permet pour tout les payement d'avoir une seul ligne alors qu'il y a plus de payement que de commandes
# transforme plusieurs ligne en plusieur colonne

payments = olist_payments.pivot_table(index='order_id', columns='payment_type', values='payment_value', fill_value=0).reset_index()
donnees = commandes.merge(payments, on="order_id")
donnees.head()

```

## All table joined to parquet

```python
donnees.to_parquet('../../_git/donnees/TP/ebrasil/donnees_all',compression='gzip', engine='pyarrow')
```
