```python
import pandas as pd, numpy as np, os, warnings, seaborn as sns
from datetime import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
font1 = fm.FontProperties(size=20)
font2 = fm.FontProperties(size=24)

# %matplotlib inline
%pylab inline

if int(str(sns.__version__).split('.')[1]) > 8 : 
    plt.style.use('seaborn-v0_8-darkgrid')
else:
    plt.style.use('seaborn-darkgrid')
    
sns.set(font_scale=2)
warnings.filterwarnings(action="ignore")

os.chdir("C:/Users/etien/Documents/CCI/_HUB/cours/2024-2025/big_data/_git/donnees")
```

```python
palette_large = [ "#030aa7", "#e50000", "#d8863b", "#005f6a", "#6b7c85", "#751973", 
            "#0485d1", "#ff7855", "#fbeeac", "#0cb577", "#95a3a6", "#c071fe", 
            "#d1e5f0", "#fddbc7", "#ffffcb", "#12e193", "#d8dcd6", "#dfc5fe", 
          ]
palette = ['#b2182b','#008827','#053061','#ffa500', '#34495e', '#e74c3c', 
          '#9b59b6', '#0165fc','#2ecc71', '#cb7723', '#89a0b0']
sns.palplot(sns.color_palette(palette))
sns.palplot(sns.color_palette(palette_large))
plt.show()
```

```python
ventes = pd.read_parquet('ecommerce/analyse_ventes.parquet', engine='pyarrow')
ventes
```

```python
ventes.info()
```

```python
ventes['time_to_approved'] = ventes['estimated_delivery'] - ventes['approved_at']
EDT_by_state = ventes.groupby('state').agg({'time_to_approved':'mean'}).reset_index()
plt.figure(figsize=(15, 8))
sns.barplot(x='state', y='time_to_approved', data=EDT_by_state, palette=palette)
plt.xticks(rotation=90)
plt.xlabel('State', fontproperties=font1)
plt.ylabel('Estimated Delivery', fontproperties=font1)
plt.title('Average Estimated Delivery by State', fontproperties=font2)
plt.gcf()
```

```python
ventes['estimated_delivery_time'] = ventes['approved_at'] - ventes['purchase_timestamp']
EDT_by_state = ventes.groupby('state').agg({'estimated_delivery_time':'mean'}).reset_index()
plt.figure(figsize=(15, 8))
sns.barplot(x='state', y='estimated_delivery_time', data=EDT_by_state, palette=palette)
plt.xticks(rotation=90)
plt.xlabel('State', fontproperties=font1)
plt.ylabel('Estimated Delivery', fontproperties=font1)
plt.title('Average Estimated Delivery by State', fontproperties=font2)
plt.gcf()
```

```python
comment_by_state = ventes.groupby('state').agg({'comment':'count'}).reset_index()

plt.figure(figsize=(15, 8))
sns.barplot(x='state', y='comment', data=comment_by_state, palette=palette)
plt.xticks(rotation=90)
plt.xlabel('State', fontproperties=font1)
plt.ylabel('Number of Comments', fontproperties=font1)
plt.title('Number of Comments by State', fontproperties=font2)
plt.gcf()
```

```python
# Group the data by 'state' and sum the sales values
sales_by_region = ventes.groupby('state')['value_credit_card'].sum().reset_index()

# Create the bar plot
plt.figure(figsize=(14, 8))
sns.barplot(data=sales_by_region, x='state', y='value_credit_card', palette=palette)
plt.title('Les ventes par rÃ©gion', fontproperties=font2)
plt.xlabel('RÃ©gion', fontproperties=font1)
plt.ylabel('Ventes (en valeur)', fontproperties=font1)
plt.xticks(rotation=45, fontproperties=font1)
plt.gcf()
```
