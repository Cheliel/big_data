```python
palette_large = [ "#030aa7", "#e50000", "#d8863b", "#005f6a", "#6b7c85", "#751973", 
            "#0485d1", "#ff7855", "#fbeeac", "#0cb577", "#95a3a6", "#c071fe", 
            "#d1e5f0", "#fddbc7", "#ffffcb", "#12e193", "#d8dcd6", "#dfc5fe", 
          ]
palette = ['#b2182b','#008827','#053061','#ffa500', '#34495e', '#e74c3c', 
          '#9b59b6', '#0165fc','#2ecc71', '#cb7723', '#89a0b0']
sns.palplot(sns.color_palette(palette))
sns.palplot(sns.color_palette(palette_large))
plt.show()
```

```python
ventes = pd.read_parquet('ecommerce/analyse_ventes.parquet', engine='pyarrow')
ventes
```

# Nombre de vente

```python
# Group the data by 'state' and sum the sales values
sales_by_region = ventes.groupby('state')['value_credit_card'].sum().reset_index()

# Create the bar plot
plt.figure(figsize=(14, 8))
sns.barplot(data=sales_by_region, x='state', y='value_credit_card', palette=palette)
plt.title('Les ventes par région', fontproperties=font2)
plt.xlabel('Région', fontproperties=font1)
plt.ylabel('Ventes (en valeur)', fontproperties=font1)
plt.xticks(rotation=45, fontproperties=font1)
plt.gcf()
```

```python
# Group by day
orders_by_day = ventes.groupby('jour').size().reset_index(name='order_count')
plt.figure(figsize=(15, 8))
sns.lineplot(x='jour', y='order_count', data=orders_by_day, color=palette[0])
plt.xticks(rotation=90)
plt.xlabel('Day', fontproperties=font1)
plt.ylabel('Number of Orders', fontproperties=font1)
plt.title('Number of Orders by Day', fontproperties=font2)
plt.show()

# Group by week
orders_by_week = ventes.groupby('semaine').size().reset_index(name='order_count')
plt.figure(figsize=(15, 8))
sns.lineplot(x='semaine', y='order_count', data=orders_by_week, color=palette[1])
plt.xticks(rotation=90)
plt.xlabel('Week', fontproperties=font1)
plt.ylabel('Number of Orders', fontproperties=font1)
plt.title('Number of Orders by Week', fontproperties=font2)
plt.show()

# Group by month
orders_by_month = ventes.groupby('annee_mois').size().reset_index(name='order_count')
plt.figure(figsize=(15, 8))
sns.lineplot(x='annee_mois', y='order_count', data=orders_by_month, color=palette[2])
plt.xticks(rotation=90)
plt.xlabel('Month', fontproperties=font1)
plt.ylabel('Number of Orders', fontproperties=font1)
plt.title('Number of Orders by Month', fontproperties=font2)
plt.show()

# Group by quarter
orders_by_quarter = ventes.groupby('annee_trimestre').size().reset_index(name='order_count')
plt.figure(figsize=(15, 8))
sns.lineplot(x='annee_trimestre', y='order_count', data=orders_by_quarter, color=palette[3])
plt.xticks(rotation=90)
plt.xlabel('Quarter', fontproperties=font1)
plt.ylabel('Number of Orders', fontproperties=font1)
plt.title('Number of Orders by Quarter', fontproperties=font2)
plt.show()

# Group by year
orders_by_year = ventes.groupby('annee').size().reset_index(name='order_count')
plt.figure(figsize=(15, 8))
sns.lineplot(x='annee', y='order_count', data=orders_by_year, color=palette[4])
plt.xticks(rotation=90)
plt.xlabel('Year', fontproperties=font1)
plt.ylabel('Number of Orders', fontproperties=font1)
plt.title('Number of Orders by Year', fontproperties=font2)
plt.show()
```

```python
ventes['time_to_approved'] = (ventes['estimated_delivery'] - ventes['approved_at']).dt.seconds / 3600
EDT_by_state = ventes.groupby('state').agg({'time_to_approved':'mean'}).reset_index()
plt.figure(figsize=(15, 8))
sns.barplot(x='state', y='time_to_approved', data=EDT_by_state, palette=palette)
plt.xticks(rotation=90)
plt.xlabel('State', fontproperties=font1)
plt.ylabel('Estimated Delivery', fontproperties=font1)
plt.title('Average Estimated Delivery by State', fontproperties=font2)
plt.gcf()
```

## Temps de livraison & temps de livraison estimé par pays

```python

ventes['jour'] = ventes['purchase_timestamp'].dt.date
SP = ventes[ventes['state'] == 'SP'].groupby('jour').agg({'estimated_delivery_time':'mean', 'real_delivery_time':'mean'}).reset_index()

plt.figure(figsize=(15, 8))
sns.lineplot(x='jour', y='estimated_delivery_time', data=SP, label='Estimated Delivery Time', color=palette[0])
sns.lineplot(x='jour', y='real_delivery_time', data=SP, label='Real Delivery Time', color=palette[1])
plt.xticks(rotation=90)
plt.xlabel('Date', fontproperties=font1)
plt.ylabel('Delivery Time (days)', fontproperties=font1)
plt.title('Estimated vs Real Delivery Time for SP', fontproperties=font2)
plt.legend()
plt.gcf()

# uncomment for all states plot

#for state in ventes['state'].unique():
#    state_data = ventes[ventes['state'] == state].groupby('jour').agg({'estimated_delivery_time':'mean', 'real_delivery_time':'mean'}).reset_index()
#    plt.figure(figsize=(15, 8))
#    sns.lineplot(x='jour', y='estimated_delivery_time', data=state_data, label='Estimated Delivery Time', color=palette[0])
#    sns.lineplot(x='jour', y='real_delivery_time', data=state_data, label='Real Delivery Time', color=palette[1])
#    plt.xticks(rotation=90)
#    plt.xlabel('Date', fontproperties=font1)
#    plt.ylabel('Delivery Time (days)', fontproperties=font1)
#    plt.title(f'Estimated vs Real Delivery Time for {state}', fontproperties=font2)
#    plt.legend()
#plt.show()
```

```python
ventes['estimated_approved_time'] = (ventes['approved_at'] - ventes['purchase_timestamp']).dt.seconds / 3600
EDT_by_state = ventes.groupby('state').agg({'estimated_approved_time':'mean'}).reset_index()
plt.figure(figsize=(15, 8))
sns.barplot(x='state', y='estimated_approved_time', data=EDT_by_state, palette=palette)
plt.xticks(rotation=90)
plt.xlabel('State', fontproperties=font1)
plt.ylabel('Estimated approved time by hour', fontproperties=font1)
plt.title('Average approved time by State', fontproperties=font2)
plt.gcf()
```

## Nombre de commande par heure pour un pays

```python
# Filtrer les données pour un pays spécifique, par exemple 'SP'
country_data = ventes[ventes['state'] == 'SP']

# Grouper les données par heure et compter le nombre de commandes
orders_by_hour = country_data.groupby('heure').size().reset_index(name='order_count')

# Créer le graphique
plt.figure(figsize=(15, 8))
sns.barplot(x='heure', y='order_count', data=orders_by_hour, palette=palette)
plt.xlabel('Heure', fontproperties=font1)
plt.ylabel('Nombre de Commandes', fontproperties=font1)
plt.title('Nombre de Commandes par Heure pour SP', fontproperties=font2)
plt.xticks(fontproperties=font1)
plt.gcf()
```

```python
# Group the data by 'jour_semaine' and 'heure' and count the number of orders
orders_by_day_hour = ventes.groupby(['jour_semaine', 'heure']).size().reset_index(name='count')

# Pivot the data to create a heatmap-friendly format
orders_pivot = orders_by_day_hour.pivot(index='jour_semaine', columns='heure', values='count')

# Define the labels for the days of the week
days_labels = ['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche']

# Plot the heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(orders_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5, cbar_kws={'label': 'Nombre de commandes'}, annot_kws={"size": 8})
plt.title('Heatmap des commandes par jour de la semaine et heure', fontproperties=font2)
plt.xlabel('Heure', fontproperties=font1)
plt.ylabel('Jour de la semaine', fontproperties=font1)
plt.yticks(ticks=np.arange(len(days_labels)) + 0.5, labels=days_labels, rotation=45, fontproperties=font1)
plt.xticks(ticks=np.arange(24) + 0.5, labels=np.arange(1, 25), fontproperties=font1)
plt.gcf()
```
